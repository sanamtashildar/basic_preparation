Here’s a Python program to rearrange the list so that it alternates between positive and negative numbers: Approach:
nums = [1, 2, 3, -4, -1, 4] o/p [1, -4, 2, -1, 3, 4]

```python
def rearrange_alternate(nums):
    # Separate positive and negative numbers
    pos = [x for x in nums if x >= 0]
    neg = [x for x in nums if x < 0]

    result = []
    i = j = 0
    turn = True  # True for positive, False for negative

    # Merge alternately
    while i < len(pos) and j < len(neg):
        if turn:
            result.append(pos[i])
            i += 1
        else:
            result.append(neg[j])
            j += 1
        turn = not turn

    # Append remaining elements
    result.extend(pos[i:])
    result.extend(neg[j:])

    return result

# Example usage
nums = [1, 2, 3, -4, -1, 4]
print(rearrange_alternate(nums))
```

# longest unique substring

```python

string = "abcpoiuabdefrjhyabhpoabc"

maxi = 0
r = 0
l = 0 
hmap_map = {}
start_index = 0   # track start of longest substring

while r < len(string):
    if string[r] in hmap_map and hmap_map[string[r]] >= l:
        l = hmap_map[string[r]] + 1   # move left pointer

    # check if this window is the largest so far
    if r - l + 1 > maxi:
        maxi = r - l + 1
        start_index = l   # store where longest substring begins

    hmap_map[string[r]] = r
    r += 1

longest_substring = string[start_index:start_index+maxi]
print(hmap_map)
print("Longest Substring:", longest_substring)
print("Length:", maxi)
```

Q Reverse sentence "Hey, how are you doing ?"

```python
sentence = "Hey, how are you doing ?"

def reverse_word(word):
    letters = [ch for ch in word if ch.isalpha()]   # collect only letters
    reversed_word = ""
    
    # build reversed word but keep punctuation
    for ch in word:
        if ch.isalpha():
            reversed_word += letters.pop()          # take from end
        else:
            reversed_word += ch                     # keep punctuation
    return reversed_word

words = sentence.split()
result = []

for word in words:
    result.append(reverse_word(word))
    # rebuild sentence
final_sentence = ""
for i in range(len(result)):
    final_sentence += result[i]
    if i != len(result) - 1:
        final_sentence += " "

print(final_sentence)
```

Q input s = "TTTTTTttttt6667aaaaaAAAA557BBBBBbbbbbb33338888000LLLLL099eeeEEE" 
output = "TaBLe"

```python
def extract_by_first_case(s: str) -> str:
    i = 0
    n = len(s)
    out = []
    while i < n:
        ch = s[i]
        if ch.isalpha():
            # start a letter-group whose identity is ch.lower()
            base = ch.lower()
            first_is_upper = ch.isupper()
            # consume the group: subsequent letters with same base (ignore case)
            j = i + 1
            while j < n and s[j].isalpha() and s[j].lower() == base:
                j += 1
            # append the base char with the case of the first char in the group
            out.append(base.upper() if first_is_upper else base)
            i = j
        else:
            # skip digits / other chars
            i += 1
    return "".join(out)
```

# Q write a function to reverse only alphabets not special character from below string
    # input_data = "Hell! wor?d" # reverse the string only alphabet

```python
def reverse_only_alphabets_inplace(s: str) -> str:
    chars = list(s)  # convert to list (mutable)
    i, j = 0, len(chars) - 1

    while i < j:
        if not chars[i].isalpha():
            i += 1
        elif not chars[j].isalpha():
            j -= 1
        else:
            # swap only alphabets
            chars[i], chars[j] = chars[j], chars[i]
            i += 1
            j -= 1

    return "".join(chars)
# Example
input_data = "Hell! wor?d"
print(reverse_only_alphabets_inplace(input_data))  # "drow! lle?H"

# Sol 2
import re

def reverse_only_alphabets2(s: str) -> str:
    # Get all alphabets
    letters = re.findall(r"[A-Za-z]", s)[::-1]  # reverse them
    it = iter(letters)  # make an iterator

    # Replace each alphabet with the next reversed one
    return re.sub(r"[A-Za-z]", lambda _: next(it), s)


# Example
input_data = "Hell! wor?d"
print(reverse_only_alphabets2(input_data))  # "drow! lle?H"

# Sol 3
def reverse_only_alphabets(s: str) -> str:
    # Extract only alphabets and reverse them
    letters = [ch for ch in s if ch.isalpha()][::-1]

    result = []
    letter_index = 0

    for ch in s:
        if ch.isalpha():
            result.append(letters[letter_index])
            letter_index += 1
        else:
            result.append(ch)

    return "".join(result)


# Example
input_data = "Hell! wor?d"
print(reverse_only_alphabets(input_data))  # "drow! lle?H"
```

# Q To convert an integer to a string without using built-in methods in Python, you can implement your own logic using arithmetic operations and character manipulation. Here's an example:
    # eg 123 -> "123"

```python

def int_to_string(num: int) -> str:
    if num == 0:
        return "0"
    
    result = []
    while num > 0:
        digit = num % 10                      # take last digit
        result.append(chr(ord('0') + digit))  # convert digit → char
        num //= 10                            # remove last digit
    
    return "".join(reversed(result))          # reverse to get correct order

# Example
print(int_to_string(123))   # "123"
print(int_to_string(4056))  # "4056"
```

# Remove duplicate from the sorted array

```python
arr = [1,1,2,2,3,3,5,5,6,8]
def removeduplicate(arr):
    i = 0
    j = i+ 1
    while j < len(arr):
        if arr[i] != arr[j]:
            i+=1
            arr[i], arr[j] = arr[j], arr[i]
        j+=1
    return i + 1
print(removeduplicate(arr))
```

# Find the second largest element form the list

```python
arr = [1,1,2,2,3,3,5,5,6,8]
def slargest(arr):
    sl = float("-inf")
    l = float("-inf")
    for i in range(len(arr)):
        if arr[i] > l:
            sl = l
            l = arr[i]
        elif arr[i] > sl and arr[i] != l:
            sl = arr[i]    
    return sl        
print(slargest(arr))
```

# Find the largest element from the list

```python

arr = [1,1,2,2,3,3,5,5,6,8]
def solution(arr):
    l = arr[0]
    for i in arr:
        if i > l:
            l = i
    return l
print(solution(arr))
```
