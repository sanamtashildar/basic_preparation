```python
# sum the number at odd position and muliply the num at even pos # result = mul/sum
nums = [24,53,62,90,36,42,81]

m = 1
for i in range(0, len(nums), 2):   # even positions (0,2,4,6)
    m *= nums[i]

s = 0
for i in range(1, len(nums), 2):   # odd positions (1,3,5)
    s += nums[i]

r = m / s
print(r)

# single loop solution
nums = [24,53,62,90,36,42,81]

m = 1
s = 0

for i in range(len(nums)):
    if i % 2 == 0:      # even index
        m *= nums[i]
    else:               # odd index
        s += nums[i]

result = m / s
print(result)
```

```python
# Serialize and Deserialize an Open File Object Using JSON

import json

# Serialize the process of converting a Python object into a format that can be stored or transmitted JSON
data = {'name': 'John', 'age': 30}
with open('data.json', 'w') as file:
    json.dump(data, file)

# Deserialize converting the stored format back into a working Python object
with open('data.json', 'r') as file:
    serialized_data = file.read()
    file.seek(0)
    loaded_data = json.load(file)

print("Type of serialized data:", type(serialized_data))
print("\nDeserialized data:", loaded_data)
print("Type of deserialized data:", type(loaded_data))
```

```python
# filter() function is used to extract elements from an iterable (like a list, tuple or set) that satisfy a given condition. 
# It works by applying a function to each element and keeping only those for which function returns True.
def starts_a(w):
    return w.startswith("a")

li = ["apple", "banana", "avocado", "cherry", "apricot"]
res = filter(starts_a, li)
print(list(res))

def even(n):
    return n % 2 == 0

a = [1, 2, 3, 4, 5, 6]
b = filter(even, a)
print(list(b)) # Convert filter object to a list
```

```python
# How do you count number of objects in Python?
# We can use a class-level static variable and increment it inside __init__() method. 
# Optionally, we can decrement it inside __del__(), but since Python uses garbage collection, __del__() 
# execution is not guaranteed immediately.
class Test:
    obj_count = 0   # class variable

    def __init__(self):
        Test.obj_count += 1

    def __del__(self):
        Test.obj_count -= 1


t = Test()
print("obj count =", Test.obj_count)   # 1

t1 = Test()
print("obj count =", Test.obj_count)   # 2

t2 = Test()
print("obj count =", Test.obj_count)   # 3

del t1
print("obj count =", Test.obj_count)   # 2
```

```python
def retry(func):
    def wrapper(*args, **kwargs):
        attempts = 3
        for i in range(attempts):
            try:
                return func(*args, **kwargs)
            except Exception as e:
                print(f"Retry {i+1} failed")
                if i == attempts - 1:
                    raise
    return wrapper
```

```python
# Make api a call, retrieve token in from it and use it in API b call
import requests
import json

# -------- API A (GET) --------
url_a = "https://api-a.com/get-token"

res = requests.get(url_a)

if res.status_code == 200:
    response_data = res.json()
    token = response_data.get("auth_token")

    if not token:
        raise Exception("Token not found in API A response")
else:
    raise Exception(f"API A failed with status code {res.status_code} - {res.text}")


# -------- Prepare Payload --------
file_location = "payload.json"

with open(file_location, "r") as f:
    payload = json.load(f)


# -------- API B (POST) --------
url_b = "https://api-b.com/create"

headers = {
    "Authorization": f"Bearer {token}",   # <-- correct way
    "Content-Type": "application/json"
}

res2 = requests.post(
    url=url_b,
    headers=headers,
    json=payload
)

print("Status:", res2.status_code)
print("Response:", res2.text)

# First API acts as authentication provider. 
# We extract token from response body or headers and pass it as Bearer token in Authorization 
# header of second API call. 
# We also validate token existence before proceeding.
```

```python
# Python code to illustrate 
# working of try()  
def divide(x, y): 
    try: 
        # Floor Division : Gives only Fractional 
        # Part as Answer 
        result = x // y 
    except ZeroDivisionError: 
        print("Sorry ! You are dividing by zero ") 
    else:
        print("Yeah ! Your answer is :", result) 
    finally:  
        # this block is always executed   
        # regardless of exception generation.  
        print('This is always executed')   

# Look at parameters and note the working of Program 
# divide(3, 2) 
divide(3, 0)
```

```python
s = "aaabbabbbbbbcccbbbcc"
o = ""
hash_map = {}
count = 1

for i in range(1, len(s)):
    if s[i] == s[i - 1]:
        count += 1
    else:
        hash_map[s[i-1]] = max(hash_map.get(s[i-1], 0), count)
        count = 1

# last character
hash_map[s[i - 1]] = max(hash_map.get(s[i - 1], 0), count)

for k, v in hash_map.items():
    o += k + str(hash_map.get(k))

print(o )
```

```python
# If you have a rating in digits (like 4.38, 3.7, etc.) and 
# want to convert it into a 5-star representation with full stars, half stars, 
# and empty stars, you can do it like this.
def star_rating(r):
    r = float(r)              # convert string to float
    rating = round(r * 2) / 2 # round to nearest 0.5
    
    full = int(rating)
    half = 1 if rating - full == 0.5 else 0
    empty = 5 - full - half

    return ["full"] * full + ["half"] * half + ["empty"] * empty


print(star_rating(".38"))
```

```python
# You are given a string.
# If the character is m → duplicate the previous character and then remove m.
# If the character is n → remove the next character and then remove n.
def process_string(s):
    result = []
    i = 0
    
    while i < len(s):
        if s[i] == 'm':
            if result:                 # duplicate previous char
                result.append(result[-1])
            i += 1                     # remove m
            
        elif s[i] == 'n':
            i += 2                     # skip n and next char
            
        else:
            result.append(s[i])
            i += 1

    return "".join(result)


print(process_string("abmcd"))
print(process_string("abncd"))
```
