How to keep API Tests idempotent?
Running the same test multiple times should produce the same result without unintended side effects.
To keep API tests idempotent, I ensure each test generates isolated test data, cleans up created resources, avoids shared state, and uses dedicated environments. I design tests to be self-contained and independent of execution order. For async systems, I implement polling instead of fixed waits. Additionally, I mock external dependencies to avoid rate limits and unintended side effects.

What if the DOM changes frequently and locators break often?
If the DOM changes frequently, I try to use stable locators like data-testid or data-qa attributes instead of dynamic IDs or classes. I avoid absolute XPath and prefer relative XPath or CSS selectors. I also use dynamic locator functions and parent-child relationships to make locators more resilient. Additionally, I work with developers to add automation-friendly attributes to reduce locator flakiness.

Q - How to handle shadow dom in Selenium?

PyShadow is a Selenium wrapper library that simplifies interaction with Shadow DOM elements. Instead of manually accessing shadow roots using JavaScript or Selenium 4 APIs, PyShadow automatically traverses shadow DOM layers and allows us to locate elements using normal CSS selectors.

```python
from selenium import webdriver
from pyshadow.main import Shadow

driver = webdriver.Chrome()
driver.get("https://example.com")

shadow = Shadow(driver)

shadow.find_element("#username").send_keys("testuser")
shadow.find_element("#password").send_keys("password")
shadow.find_element("#login").click()
```

Q - How to handle shadow dom in playwright?
In Playwright, shadow DOM is handled automatically. We can access elements inside shadow roots using normal locators or the >> selector syntax. For nested shadow DOMs we chain selectors like parent >> child >> element. In rare cases we can use page.evaluate() to access the shadowRoot via JavaScript.
